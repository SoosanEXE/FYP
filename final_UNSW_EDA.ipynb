```python
import pandas as pd
df_unsw = pd.read_csv("FYP/unsw.csv")
```

```python
df_unsw.drop('label',axis =1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def col_countplot(col, train_data=df_unsw):
    fig, ax = plt.subplots(figsize=(12,9))
    sns.set_style('whitegrid')
    # countplot of the given column
    ax = sns.countplot(x=col, data=train_data)
    ax.legend(loc="upper right", labels=('normal', 'attack'))
    ax.set_title("train data")
    plt.xticks(rotation=45)
    plt.show()
```

```python
col_countplot('attack_cat')
```

```python
df_unsw['attack_cat'].value_counts()
```

```python
corr = df_unsw.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
def corr(col1, col2="label", df=df_unsw):
    return df[[col1, col2]].corr().iloc[0,1]
```

```python
method = "pearson"
# correlation matrix
corr_mat = df_unsw.corr(method=method)
plt.figure(figsize=(12,12)) 
sns.heatmap(corr_mat, square=True)
plt.show()
```
